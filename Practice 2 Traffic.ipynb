```python
days = ['Mon' , 'Tues' , 'Wed' , 'Thurs' , 'Fri']

traffic_per_day = {}

for i, day in zip(range(0,135,27),days):
    traffic_per_day[day] = traffic.iloc[i:i+27]
```

```python
traffic_per_day['Tues']
```

```python
for day in days:
    plt.plot(traffic_per_day[day]['Hour (Coded)'], traffic_per_day[day]['Slowness in traffic (%)'] , label =day)
    
    plt.legend()
             
plt.show()
```

```python
plt.figure(figsize= (10,12))

for i, day in zip(range(1,6),days):
    plt.plot(traffic_per_day[day]['Hour (Coded)'] , traffic_per_day[day]['Slowness in traffic (%)'])
    plt.subplot(3,2,i)
    plt.ylim([0,25])
    plt.xlim([0,30])
    plt.title(day)
    
plt.subplot(3,2,6)

for day in days:
    plt.plot(traffic_per_day[day]['Hour (Coded)'] , traffic_per_day[day]['Slowness in traffic (%)'])
    plt.ylim([0,25])
    plt.xlim([0,30])
    
plt.legend()
plt.show()
```
